```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from itertools import zip_longest
import re
import numpy as np
```

# Web Scraping - Yelp

```python
url = 'https://www.yelp.com/biz/kaito-m%C3%BCnchen?osq=kaito'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
```

## Scrape Data

### Comment time

```python
comment_time = soup.find_all('span', class_ = 'y-css-1d8mpv1')
comment_time
```

### Username

```python
username = soup.find_all('a', class_ = 'y-css-1ijjqcc')
username
```

### Comment text

```python
comment = soup.find_all('span', class_ = 'raw__09f24__T4Ezm')
comment
```

### Rating out of 5 stars

```python
rating = soup.find_all('div', class_ = 'y-css-dnttlc')
rating
```

## Extract Text from HTML

```python
data = []
date_pattern = re.compile(r'^[A-Za-z]{3} \d{1,2}, \d{4}$') # Validate "Month Day, Year" format.
comment_time_text = [
    time.text.strip() for time in comment_time
    if time.text.strip() and date_pattern.match(time.text.strip()) # Observe the special pattern.
]
username_text = [
    user.text for user in username
    if user.get('href') and user['href'].startswith('/user_details?userid=') # Observe the special labels.
]
comments_text = [
    comment.text.strip() for comment in comment
    if comment.get('lang') == 'en' # Observe the special conditions.
]
rating_text = [
    rating.get('aria-label').split()[0]
    for rating in rating if rating.get('aria-label') # Observe the special labels
]
data = {'Time': comment_time_text, 'Username': username_text, 'Comments': comments_text, 'Rating': rating_text}
aligned_data = list(zip_longest(comment_time_text, username_text, comments_text, rating_text, fillvalue=''))
df = pd.DataFrame(aligned_data, columns=['Time', 'Username', 'Comments', 'Rating out of 5 Stars'])
df = df.replace(r'^\s*$', np.nan, regex=True)
df = df.fillna({
    'Time': 'N/A',
    'Username': 'Anonymous',
    'Comments': 'No comments provided',
    'Rating out of 5 Stars': 'N/A'
})
df.index = range(1, len(df) + 1)
df.to_csv('Kaito_Reivew.csv', index=False)  
df
```
